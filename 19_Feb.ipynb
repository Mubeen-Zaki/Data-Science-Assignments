##### 1) Flask: It is a python framework or library which helps in accessing python applications/functions running on the server machine from a client machine (anywhere). It acts as an API and lets the client access the services through a URL. It is used in web dev. 
##### Advantages:
1. Simple
2. Lightweight
3. Reusable
4. Accessible from anywhere, etc.
